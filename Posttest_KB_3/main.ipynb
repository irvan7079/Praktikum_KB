# Import Library Pandas, Matplotlib, dan Seaborn

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Menginput File CSV Menggunakan Pandas

```python
data = pd.read_csv('Wine_Quality_Dataset.csv')
```

# 1. Grafik Batang (Menggunakan Matplotlib)

```python
data = data.groupby('is_red').mean() # Ini buat mengelompokkan dataset menjadi 2 bagian, yaitu 0 (anggur biasa) dan 1 (anggur merah)
plt.figure(figsize=(2, 5)) # Ini buat mengatur panjang dan lebar pada grafik
x = data.index # Ini buat menampung nilai pengelompokan tadi ke dalam variabel x
y = data.fixed_acidity # Ini buat menampung nilai asam tetap ke dalam variabel y
labels = ['Non-Red Wine', 'Red Wine'] # Ini buat menampung nama label yang akan ditampilkan
plt.bar(x,y, color=['green', 'red'], label=labels) # Ini untuk membuat isi dari grafik batang
plt.title("Rata - Rata Jumlah Asam Tetap pada Anggur") # Ini buat judul grafik
plt.xlabel('Jenis Anggur') # Ini buat menamakan label yang terdapat pada koordinat X (Horizontal)
plt.ylabel('Rata - Rata nilai Fixed Acidity') # Ini buat menamakan label yang terdapat pada koordinat y (Vertikal)
plt.legend() # Ini untuk menampilkan nama label nya
plt.show() # Ini untuk menampilkan Grafik
```

#### Kesimpulan Grafik Batang: Jenis anggur merah memiliki rata-rata jumlah asam tetap yang lebih tinggi daripada anggur biasa

# 2. Grafik Histogram (Menggunakan Matplotlib)

```python
plt.hist(data['citric_acid'], bins=30, color='red')  # Ini untuk membuat isi dari grafik Histogram
plt.title("Histogram Jumlah Asam Nitrat Pada Anggur") # Ini buat judul grafik
plt.xlabel('Jumlah Asam Nitrat') # Ini buat menamakan label yang terdapat pada koordinat X (Horizontal)
plt.ylabel('Frekuensi Data') # Ini buat menamakan label yang terdapat pada koordinat y (Vertikal)
plt.show() # Ini untuk menampilkan Grafik
```

#### Kesimpulan Grafik Histogram: Sebagian besar Anggur memiliki jumlah asam nitrat di skala 0,25 sampai 0,35

# 3. Grafik Pie Chart (Menggunakan Matplotlib)

```python
pc_wine = data.is_red.value_counts() # Variable untuk menghitung jumlah data pada label is_red yang memiliki nilai yang sama
pc_wine # Ini untuk menampilkan isi variable nya
```

```python
labels = ['Non-Red Wine', 'Red Wine'] # Ini buat menampung nama label yang akan ditampilkan
plt.pie(pc_wine, labels=labels, autopct="%1.1f%%") # Ini untuk membuat isi dari grafik Pie Chart
plt.title('Jenis Anggur') # Ini buat judul grafik
plt.show() # Ini untuk menampilkan Grafik
```

#### Kesimpulan Grafik Pie Chart: Sebagian besar jenis anggur dalam dataset saya bukan anggur merah (Non-Red Wine)

# 4. Grafik Scatter Plot (Menggunakan Matplotlib)

```python
plt.figure(figsize=(20, 10)) # Ini buat mengatur panjang dan lebar pada grafik
is_red = data.loc[data.is_red == 1] # Variable ini buat menampung jumlah anggur merah
not_red = data.loc[data.is_red == 0] # Variable ini buat menampung jumlah anggur biasa

plt.scatter(is_red.sulphates, is_red.alcohol, c='red', label='Red Wine')  # Ini untuk membuat isi dari grafik Scatter Plot yang berupa titik merah
plt.scatter(not_red.sulphates, not_red.alcohol, c='green', label='Non-Red Wine') # Ini untuk membuat isi dari grafik Scatter Plot yang berupa titik hijau

plt.title('Klasifikasi Jenis Anggur') # Ini buat judul grafik
plt.xlabel('Tingkat Sulfat') # Ini buat menamakan label yang terdapat pada koordinat X (Horizontal)
plt.ylabel('Tingkat Alkohol') # Ini buat menamakan label yang terdapat pada koordinat y (Vertikal)
plt.legend() # Ini untuk menampilkan nama label nya
plt.show() # Ini untuk menampilkan Grafik
```

#### Kesimpulan Grafik Scatter Plot: Sebagian besar jenis anggur merah memiliki tingkat sulfat lebih tinggi daripada jenis anggur biasa. Untuk tingkat alkohol dalam anggur tidak terlalu mempengaruhi jenis dari anggur tersebut.

# 5. Grafik Box Plot (Menggunakan Seaborn)

```python
plt.figure(figsize=(15, 5)) # Ini buat mengatur panjang dan lebar grafik
sns.boxplot(x = data['pH']) # Ini untuk membuat isi dari grafik Box Plot
plt.title("Nilai pH Yang Terkandung Dalam Anggur") # Ini buat judul grafik
plt.show() # Ini untuk menampilkan Grafik
```

#### Kesimpulan Grafik Box Plot: Sebagian besar anggur hanya memiliki tingkat pH dari skala 3,1 sampai 3,34

# 6. Grafik Heat Map (Menggunakan Seaborn)

```python
corr = data.corr() # Untuk membuat variabel corr yang akan digunakan dalam pembuatan grafik Heatmap
sns.heatmap(corr, cmap='coolwarm', annot=True, center=0) # Untuk membuat grafik Heatmap
plt.title("Heatmap Spesifikasi Anggur") # Ini buat judul grafik
plt.show() # Ini untuk menampilkan Grafik
```

#### Kesimpulan Grafik Heat Map: Komponen-komponen yang ada di dalam anggur rata-rata memiliki pengaruh yang kurang lebih sama dalam menentukan jenis dari anggur sehingga sebagian besar warna dalam grafik Heatmap memiliki warna yang tidak terlalu panas (Merah Tua) dan tidak terlalu dingin (Biru Tua)
