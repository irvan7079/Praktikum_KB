```python
import pandas as pd
```

# Menginput File CSV

```python
data = pd.read_csv('Wine_Quality_Dataset.csv')
```

# 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin

```python
# Informasi jumlah record
num_records = len(data)

# Informasi jumlah attribute
num_attributes = len(data.columns)

# Informasi nama attribute
attribute_names = data.columns.tolist()

# Informasi jumlah attribute dengan tipe angka
numeric_attributes = data.select_dtypes(include=['number']).columns.tolist()
num_numeric_attributes = len(numeric_attributes)

# Membuat dictionary dengan informasi yang ada diatas
dataset_info = {
    'Jumlah Record': num_records,
    'Jumlah Attribute': num_attributes,
    'Nama Attribute': attribute_names,
    'Jumlah Attribute dengan Tipe Angka': num_numeric_attributes,
}

print("===========================")
print("=== Dataset Information ===")
print("===========================")

# Menampilkan isi dictionary
for key, value in dataset_info.items():
    print(f"{key}: {value}")
```

# 5. Tampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama)

```python
# Mengidentifikasi atribut yang memiliki tipe 'object' atau string
object_attributes = data.select_dtypes(include=['object'])

# Menghitung jumlah label unik untuk setiap atribut 'object'
label_counts = {}
for column in object_attributes.columns:
    label_counts[column] = len(data[column].unique())

# Tampilkan hasilnya
for attribute, count in label_counts.items():
    print(f"Atribut: {attribute}, Jumlah Label Unik: {count}")
```
